```python
# Insertion sort algorithm to sort the 'Value' column in ascending order
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Extract the 'Value' column as a list
values = health['Value'].tolist()

# Sort the 'Value' column using insertion sort
sorted_values = insertion_sort(values)

# Assign the sorted values back to the 'Value' column
health['Sorted Value'] = sorted_values

print("\nDataset with sorted 'Value' column:")
print(health.head())

# To show the entire sorted dataset
print("\nEntire sorted dataset:")
print(health)
```

## Explanation of Search Algorithms: Linear Search vs Binary Search
## Linear Search:

- **Performance**: Linear search has a time complexity of 

(

)
O(n), where 

n is the number of elements in the list. This means that in the worst case, it has to check every element in the list to find the target.
- **Use Case**: Linear search can be used on unsorted lists. It is straightforward and easy to implement.
- **Speed**: It performs slower on large datasets because it needs to iterate through each element until it finds the target or reaches the end of the list.

## Binary Search:

- **Performance**: Binary search has a time complexity of 

(
log


)
O(logn), where 

n is the number of elements in the list. This is much faster than linear search for large datasets.
- **Use Case**: Binary search can only be used on sorted lists. It works by repeatedly dividing the search interval in half.
Speed: It performs faster on large datasets because it reduces the search space by half at each step.

## Python Code for Searching a Specific Element in the Dataset
We will write Python code to search for a specific year in the Year column of the dataset using both linear search and binary search algorithms.

```python
# Linear search algorithm
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Binary search algorithm
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Extract the 'Year' column as a list
years = health['Year'].tolist()

# Sort the years for binary search
sorted_years = sorted(years)

# Define the target year to search for
target_year = 2000

# Perform linear search
linear_result = linear_search(years, target_year)
print(f"Linear Search: Year {target_year} found at index {linear_result}")

# Perform binary search
binary_result = binary_search(sorted_years, target_year)
print(f"Binary Search: Year {target_year} found at index {binary_result} in sorted list")

# Verify the binary search result in the original list
if binary_result != -1:
    binary_result_in_original = years.index(sorted_years[binary_result])
    print(f"Binary Search: Year {target_year} found at index {binary_result_in_original} in original list")
else:
    print(f"Binary Search: Year {target_year} not found in the original list")
```

### Complexity Analysis: Insertion Sort Algorithm

**Insertion Sort** is the sorting algorithm we implemented to sort the `Year` column in the dataset.

#### Time Complexity

1. **Best Case**:
   - The best case occurs when the input list is already sorted.
   - In this scenario, the inner while loop condition `key < arr[j]` fails immediately for every element.
   - The algorithm only performs the outer for loop operations, which are \(n-1\) comparisons.
   - **Best Case Time Complexity**: \(O(n)\)

2. **Average Case**:
   - In the average case, each element in the list is compared with half of the sorted portion of the list on average.
   - The inner while loop executes for about half of the elements in each iteration of the outer for loop.
   - The average number of comparisons and shifts is \(\frac{n(n-1)}{4}\).
   - **Average Case Time Complexity**: \(O(n^2)\)

3. **Worst Case**:
   - The worst case occurs when the input list is sorted in reverse order.
   - In this scenario, each element must be compared and shifted with all previously sorted elements.
   - The total number of comparisons and shifts is \(\frac{n(n-1)}{2}\).
   - **Worst Case Time Complexity**: \(O(n^2)\)

#### Space Complexity

- **Space Complexity**:
  - Insertion sort is an in-place sorting algorithm, meaning it requires a constant amount of additional memory space.
  - Only a few additional variables are used, regardless of the size of the input list.
  - **Space Complexity**: \(O(1)\)

### Efficiency

- **Efficiency**:
  - **Insertion Sort** is efficient for small or nearly sorted datasets due to its simple implementation and low overhead.
  - However, it becomes inefficient for larger datasets due to its quadratic time complexity in the average and worst cases.
  - Its space efficiency is high, as it operates in-place without requiring additional memory proportional to the input size.

### Summary

- **Time Complexity**: Best case \(O(n)\), Average case \(O(n^2)\), Worst case \(O(n^2)\)
- **Space Complexity**: \(O(1)\)

Insertion Sort is best suited for small or nearly sorted datasets but is not optimal for large datasets due to its \(O(n^2)\) time complexity in average and worst cases.
